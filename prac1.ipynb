```python
import matplotlib.pyplot as plt
from abc import ABC, abstractmethod

class BioData(ABC):
    def __init__(self, name, data):
        self.name = name
        self.data = data

    @abstractmethod
    def analyze(self):
        pass

    @abstractmethod
    def visualize(self):
        pass

class ProteinSequence(BioData):
    def __init__(self, name, data):
        super().__init__(name, data)

    def analyze(self):
        print(set(self.data))  # Prints the unique amino acids
        return {"length": len(self.data), "unique_amino_acids": len(set(self.data))}

    def visualize(self):
        amino_acid_counts = {aa: self.data.count(aa) for aa in set(self.data)}
        plt.bar(amino_acid_counts.keys(), amino_acid_counts.values())
        plt.title(f"Visualization of {self.name}")
        plt.show()

class DNASequence(BioData):
    def __init__(self, name, data):
        super().__init__(name, data)

    def analyze(self):
        print(set(self.data))  # Prints the unique nucleotides
        return {"length": len(self.data), "unique_nucleotides": len(set(self.data))}

    def visualize(self):
        nucleotide_counts = {nt: self.data.count(nt) for nt in set(self.data)}
        plt.bar(nucleotide_counts.keys(), nucleotide_counts.values())
        plt.title(f"Visualization of {self.name}")
        plt.show()



def main():
    while True:
        print("Menu:")
        print("1. Analyze Protein Sequence")
        print("2. Analyze DNA Sequence")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter the protein name: ")
            sequence = input("Enter the protein sequence: ")
            protein = ProteinSequence(name, sequence)
            print(protein.analyze())
            protein.visualize()

        elif choice == "2":
            name = input("Enter the gene name: ")
            sequence = input("Enter the DNA sequence: ")
            dna = DNASequence(name, sequence)
            print(dna.analyze())
            dna.visualize()

        elif choice == "3":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
